```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter
from scipy import stats

%config IPCompleter.greedy=True
```

## Data Retrieval

```python
DIRECTORY = '/Users/diyahirc/Documents/BCM/genesets' # Change to your path
```

```python
geneSetList = []

for fileName in os.listdir(DIRECTORY):
    tempList = []
    tempPath = os.path.join(DIRECTORY, fileName)
    tempFile = open(tempPath,'r')
    print(tempPath)
    
    tempLines = tempFile.readlines()
    count = 0
    for line in tempLines:
        line = line.lower()
        line = line.strip('\n') 
        tempGeneSet = line.split('\t')[2:] # first two entries are 'geneSet name' and 'discription'   
        if len(tempGeneSet) > 1:           # not interested in single gene'd gene sets
            geneSetList.append(tempGeneSet)
```

## Exploratory Data Analysis

```python
def uniqueGenesInList(geneList):
    uniqueGenes = Counter()
    for geneSet in geneList:
        for gene in geneSet:
            uniqueGenes[gene] +=1
    return [uniqueGenes,len(uniqueGenes)]
```

```python
TOTALGENESETCOUNT = len(geneSetList)

genesCount = []
geneSetCount = 0
for geneSet in geneSetList: 
    currentLength = len(geneSet)
    geneSetCount += 1
    genesCount.append(len(geneSet))

TOTALUNIQUEGENESCOUNT = uniqueGenesInList(geneSetList)
print('Unique number of genes: {numUnique}'.format(numUnique = TOTALUNIQUEGENESCOUNT))
stats.describe(genesCount)
```

```python
testLengths = [18000,1000,500,100,50,20]
numTests = len(testLengths)
plt.figure(figsize=(10,22))
for index,testLength in enumerate(testLengths):
    genesCount = []
    geneSetCount = 0
    for geneSet in geneSetList: 
        currentLength = len(geneSet)
        if currentLength < testLength:
            geneSetCount += 1
            genesCount.append(len(geneSet))
    plt.subplot('{total}1{place}'.format(place=index+1,total=numTests))
    plt.hist(genesCount,bins=10)
    plt.title('N={total} -- {percent}% of total -- GeneSets of lengths < {count}'.format(total=geneSetCount,
                                                                                      count=testLength,
                                                                                         percent=np.round(100*geneSetCount/TOTALGENESETCOUNT)))
plt.show()
```

### Ploting Affect of Setting a Maximum Set Lenght

```python
geneSetLengths = np.linspace(2,1000,40)
uniqueGenesCount = []
numberOfSetsCount = []
for maxGeneSetLength in geneSetLengths:
    genesCount = []
    filteredGeneSetList = []
    geneSetCount = 0
    for geneSet in geneSetList: 
        currentLength = len(geneSet)
        if currentLength < maxGeneSetLength:
            geneSetCount += 1
            filteredGeneSetList.append(geneSet)
    numberOfSetsCount.append(geneSetCount/TOTALGENESETCOUNT)
    uniqueGenesCount.append(uniqueGenesInList(filteredGeneSetList)/TOTALUNIQUEGENESCOUNT)
```

```python
plt.figure(figsize=(10,5))
plt.plot(geneSetLengths,numberOfSetsCount, label='Percent of Total GeneSets')
plt.plot(geneSetLengths,uniqueGenesCount, label='Percent of Total Unique GeneSets')
plt.legend()
plt.xlabel('Max Gene Set Length')
plt.ylabel('Percentage')
plt.title('Data Completeness by Setting Max Length of Gene Set')
plt.show()
```

```python
MAXSETLENGTH = 100

genesCount = []
filteredGeneSetList = []
geneSetCount = 0
for geneSet in geneSetList: 
    currentLength = len(geneSet)
    if currentLength < MAXSETLENGTH:
        geneSetCount += 1
        filteredGeneSetList.append(geneSet)
        genesCount.append(len(geneSet))
print(stats.describe(genesCount))

plt.hist(genesCount,bins=10)
plt.title('{percent}% of total -- GeneSets of lengths < {count}'.format(total=geneSetCount,
                                                                        count=MAXSETLENGTH,
                                                                        percent=np.round(100*geneSetCount/TOTALGENESETCOUNT)))
plt.show()

NUMUNIQUE = uniqueGenesInList(filteredGeneSetList)
print('Unique number of genes: {numUnique} -- {percent}%'.format(percent=np.round(100*NUMUNIQUE/TOTALUNIQUEGENESCOUNT),
                                                                numUnique = NUMUNIQUE ))
```
